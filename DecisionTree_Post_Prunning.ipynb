```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris1=load_iris()
```

```python
#Multiclass Classification Problem
iris1.data
```

```python
iris1.target
#target --> output feature
```

```python
#Other way --> Display beautifully ==> Seperate from sklearn
import seaborn as sns
df=sns.load_dataset("iris") # putting input(output) features to dataframe
```

```python
df.head()
```

```python
#independent features and dependent features
X=df.iloc[:,:-1] # removing species using -1
y = iris1.target
```

```python
X,y
```

```python
#Train test split ---> search and copy paste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
#Sklearn Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#iF NOTHING given --> By default all paramter selected and Post Prunning
treeModel = DecisionTreeClassifier()
```

```python
#training
treeModel.fit (X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10)) #size of figure
tree.plot_tree(treeModel,filled=True)
#creates CART tree --> Every node have 2 CHild only
#By defualt Criterion --> Gini (0 to 0.5 value of every Node)
#values[v1(first category),v2 (2nd category),v3(3rd category)]--> 
#ROOT --> X[3] == Pteal_length i.e. Highest Information Gain
#gini = 0.0 ==> Leaf node and 0.5 --> Impure
```

```python
#prediction
y_pred=treeModel.predict(X_test)
```

```python
y_pred
```

```python
#To check accuracy
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```

#    Using Post Prunning --> Max Depth

```python
#Causes overfitting ---> All leaf nodes calculated
# We split upto level 2 (0,1,2) ===> 34 >> 4 and 1 << 30

treeModel = DecisionTreeClassifier(max_depth=2)
treeModel.fit (X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeModel,filled=True)
```

```python
#prediction
y_pred=treeModel.predict(X_test)
```

```python
y_pred
```

```python
#To check accuracy
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
